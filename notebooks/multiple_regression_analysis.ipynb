```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

dt = pd.read_csv('prefecture_multivariate_regression.csv', thousands=',')

# 先頭列（都道府県名）を除く
X = dt.iloc[:, 1:-1].values
y = dt.iloc[:, -1].values

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
y = imputer.fit_transform(y.reshape(-1, 1)).ravel()

#P値を求めるためのライブラリをインストール
!pip install statsmodels
import statsmodels.api as sm

#p値を求めて確認する
# 切片を追加（statsmodelsは自動で追加しない）
X_opt = sm.add_constant(X)

# 回帰モデルを作成
model = sm.OLS(y, X_opt).fit()

# 結果を表示
print(model.summary())

#どの列に文字があるのか確認
print(dt.dtypes)

# X から x1, x2, x7 のみ取り出す
X_selected = X[:, [0, 1, 6]]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size = 0.2, random_state = 0)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)

y_pred = reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))



```
